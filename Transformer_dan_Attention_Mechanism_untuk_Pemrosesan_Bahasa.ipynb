# **Import Library**

```python
from transformers import BertTokenizer, BertModel
import torch
```

# **Load tokenizer dan model pretrained**

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

# **Tokenisasi kalimat**

```python
# Pasangan A: Makna Mirip (Positif/Negatif)
kalimat_1A = "The project was an outstanding success." # Sangat positif
kalimat_2A = "The project resulted in complete failure." # Sangat negatif, tetapi sama-sama bicara 'project'

# Pasangan B: Makna Tidak Mirip
kalimat_1B = "A computer processes data efficiently." # Teknologi
kalimat_2B = "The apple pie smelled delicious." # Makanan
```

```python
def get_sentence_embedding(text):
    """Tokenisasi, forward pass, dan ambil embedding [CLS]."""
    # Tokenisasi
    inputs = tokenizer(text, return_tensors="pt")
    # Forward pass
    with torch.no_grad():
        outputs = model(**inputs)
    # Ambil embedding [CLS] (token pertama, index 0)
    # Shape: [batch_size, sequence_length, hidden_dim] -> [1, hidden_dim]
    cls_embedding = outputs.last_hidden_state[:, 0, :]
    return cls_embedding
```

# **Perhitungan Cosine Similarity**

```python
from torch.nn.functional import cosine_similarity

# Pasangan A (Mirip)
embed_1A = get_sentence_embedding(kalimat_1A)
embed_2A = get_sentence_embedding(kalimat_2A)
cos_sim_A = cosine_similarity(embed_1A, embed_2A).item()

# Pasangan B (Tidak Mirip)
embed_1B = get_sentence_embedding(kalimat_1B)
embed_2B = get_sentence_embedding(kalimat_2B)
cos_sim_B = cosine_similarity(embed_1B, embed_2B).item()
```

# **Output Hasil**

```python
print(f"Cosine Similarity (A): {cos_sim_A:.4f}")
print(f"Cosine Similarity (B): {cos_sim_B:.4f}")
```
